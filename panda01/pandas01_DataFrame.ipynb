# DataFremeの使い方

DataFrameはラベル付きの2次元の配列でいろいろな型を格納することができる


```python
import pandas as pd

a = pd.DataFrame([[80,100,20,40,"合格"],
                  [100,20,10,20,"合格"],
                  [20,50,100,0,"合格"],
                  [70,33,21,74,"不合格"],
                  [35,55,87,160,"不合格"],
                  [12,35,1,0,"不合格"]])
a


```

```python
a.index = ["ken","taro","hanako","jin","sachiko","yoko"] #行の名前を追加
a.columns = ["japanese","English","Math","Art","Result"]#列の名前を追加
a


```

```python
print("列、行",a.shape) #行数と列数を表示


```

```python
a.head()#最初の5行表示
```

```python

a.tail()#最後の5行を表示

```

```python
a.describe()#基本的な統計量
```

```python
tr = a.loc["taro",:]#taroの行を取り出す
tr
```

```python
ma = a.loc[:,"English"] #Englishの列を取り出す
ma
```

```python
r = a.iloc[1:4,:2]#行1−3、列0−1
r
```

```python
a.loc["shiro"] = pd.Series([10,20,30,40,"正解"], index=["japanese","English","Math","Art","Result"], name="shiro") #Seriesを行として追加
a
```

```python
a["Science"] = [0,1,2,3,4,5,6]# 列をリストとして追加
a
```

```python
#sort_valuesメゾットでソートをすることができる

a.sort_values(by="Math",ascending=False) #False数が大きい方、Ture数が小さい順にソートする
```
